```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('AIOE_DataAppendix.xlsx')
```

```python
# Set display options to show all columns and rows without truncation
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)     # Show all rows
pd.set_option('display.width', None)        # Width of the display in characters
pd.set_option('display.max_colwidth', None) # Show full content of each column

# Display the first 20 rows of the data
print(df.head(20))
```

```python
# List all sheet names in the Excel file
print("Available sheets in the Excel file:")
xl = pd.ExcelFile('AIOE_DataAppendix.xlsx')
print(xl.sheet_names)

# Read the specific sheet (for example, sheet 'A' for AIOE Scores)
df = pd.read_excel('AIOE_DataAppendix.xlsx', sheet_name='Appendix A')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Display the first 20 rows
print("\nFirst 20 rows of the selected sheet:")
print(df.head(20))
```
